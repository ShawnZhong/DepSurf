```python
from depsurf import VERSION_DEFAULT, prep

prep(VERSION_DEFAULT)
```

The following cell batch processes all the kernels. 

Please have more than 20GB free space. Depending on the number of cores, it will take about 10 minutes to finish.

```python
import multiprocessing as mp

from depsurf import VersionGroup, prep

versions = VersionGroup.ALL.versions
num_cores = min(mp.cpu_count(), len(versions))
# num_cores = 1

with mp.Pool(num_cores, maxtasksperchild=1) as pool:
    pool.map(prep, versions)
```
