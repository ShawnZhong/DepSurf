```python
from depsurf import DBGSYM_PATH

import pickle

with open(DBGSYM_PATH / "subroutine_info.pkl", "rb") as f:
    subroutine_info = pickle.load(f)
```

```python
dup = [(name, v) for name, v in subroutine_info.items() if len(v) > 1]
```

```python
from collections import defaultdict

sizes = defaultdict(int)
for name, v in sorted(dup, key=lambda x: -len(x[1])):
    sizes[len(v)] += 1
    print(name)
    for location, info in v.items():
        print(f"\t{location:40}")
```

```python
import pandas as pd

df = pd.DataFrame.from_dict(sizes, orient="index")

df
```
