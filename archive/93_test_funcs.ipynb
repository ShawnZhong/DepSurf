```python
from depsurf import FunctionRecorder, VERSION_DEFAULT


def cus_mapper(cus):
    return [
        next(cu for cu in cus if cu.get_top_DIE().get_full_path().endswith("main.c"))
    ]

    result = []
    for cu in cus:
        path = cu.get_top_DIE().get_full_path()
        if path.endswith("kernel/cpu.c"):
            result.append(cu)
        if path.endswith("arch/x86/kernel/smpboot.c"):
            result.append(cu)
        if len(result) == 2:
            break

    return result


recorder = FunctionRecorder.from_path(
    path=VERSION_DEFAULT.vmlinux_path,
    cus_mapper=cus_mapper,
    debug=True,
)

list(recorder.iter_funcs())
```
