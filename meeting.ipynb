# Updates on paper

Big picture of eBPF programs: see Table 1



- eBPF programs has different types defined by the kernel (e.g., kprobe, tracepoint, XDP)

- Different type of eBPF programs has different hooks to attach to
    - kprobe are attached to kernel functions

    - XDP are attached to network interface

- Different type of eBPF programs has different capabilities via eBPF helper functions
    - "each program  type  can  only  call  a  subset  of  those [eBPF] helpers" -- `man bpf-helpers`

    - e.g., kprobe cannot call helper functions to redirect packets.

    - There exists a base set of helper functions available to all eBPF programs

    - The helper functions to access kernel data structures are in the base set (`bpf_probe_read`)


On the stability of eBPF programs

From the [kernel docs](https://docs.kernel.org/bpf/bpf_design_QA.html#q-does-bpf-have-a-stable-abi):

- Q: Does BPF have a stable ABI?
  
  A: However there is one specific exception to tracing programs which are using helpers like bpf_probe_read() to walk kernel internal data structures and compile with kernel internal headers. Both of these kernel internals are subject to change and can break with newer kernels such that the program needs to be adapted accordingly.


- Q: Are tracepoints part of the stable ABI?

  A: NO. Tracepoints are tied to internal implementation details hence they are subject to change and can break with newer ker

- Q: Are places where kprobes can attach part of the stable ABI?
  
  A: NO. The places to which kprobes can attach are internal implementation details, which means that they are subject to change and can break with newer kernels. BPF programs need to change accordingly when this happens.


# Study on Linux source code

Methodology:

1. Find all supported Linux kernel by a given distribution
   (e.g., Ubuntu 20.04 supports Linux 5.4, 5.8, 5.11, 5.13, 5.15)

2. Download the compiled kernel image (`vmlinux`) 

3. Extract the debug info in BTF format (BPF Type Format), which is a compact representation of the kernel data structures consumed by eBPF loader

4. Normalize the BTF data structure to a common format

5. Compare the normalized BTF data structure across different kernel versions
