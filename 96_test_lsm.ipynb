```python
from depsurf import DepKind, VersionGroup, gen_report
from pathlib import Path

deps = sorted(set(DepKind.LSM(f) for v in VersionGroup.REGULAR for f in v.img.lsm_hooks))

data = gen_report(deps, VersionGroup.REGULAR, Path("/dev/null"))
```

```python
import pandas as pd

from utils import *

results = {}
for dep, report in data.items():
    result = {}
    for (group, version), issues in report.items():
        result[(group.name, version)] = issues.get_symbol(emoji=True)
    results[dep.name] = result

pd.DataFrame(results).T
```
