```python
from depsurf.output import load_df

df_raw = load_df("config/summary")
```

```python
from depsurf import ChangeEnum

d = {
    ChangeEnum.ADD: "$+$",
    ChangeEnum.REMOVE: "$-$",
    ChangeEnum.CHANGE: "$\Delta$",
    "New": "$\Sigma$",
}

row_selector = df_raw.index.get_level_values(-1).isin(d.keys())
col_selector = [(t, n) for (t, n) in df_raw.columns if "Oracle" not in n]


df = df_raw[col_selector].loc[row_selector]
df.index = df.index.map(lambda x: (x[0], d[x[1]]))
df.columns = df.columns.map(lambda x: (x[0], x[1].split("â†’")[-1]))
df
```
